```python
import numpy as np
from sklearn.model_selection import train_test_split
import os
import glob
import DataAug_misc as da
import librosa
```

```python
data =  np.load('Train.npz')
data = data['arr_0']
```

```python
labels =  np.load('Train_label.npz')
labels = labels['arr_0']
```

```python
data.shape
```

```python
cyan_paths = glob.glob(os.path.join('sig_noise_data_3s/Signal/Cyanistes caeruleus/', "*.wav"))
emb_paths = glob.glob(os.path.join('sig_noise_data_3s/Signal/Emberiza citrinella/', "*.wav"))
fring_paths = glob.glob(os.path.join('sig_noise_data_3s/Signal/Fringilla coelebs/', "*.wav"))
parus_paths = glob.glob(os.path.join('sig_noise_data_3s/Signal/Parus major/', "*.wav"))
pass_paths = glob.glob(os.path.join('sig_noise_data_3s/Signal/Passer montanus/', "*.wav"))
```

```python
for path in cyan_paths:
    wav_file, sr = librosa.load(path)
    mfcc = da.wav_to_mfcc(np.array(wav_file, dtype=float))
    
    data = append(data, mfcc)
    print(data.shape)
    
    labels = append(labels, np.array([1, 0, 0, 0, 0]))
    print(labels.shape)
    
for path in emb_paths:
    wav_file, sr = librosa.load(path)
    mfcc = da.wav_to_mfcc(np.array(wav_file, dtype=float))
    
    data = append(data, mfcc)
    print(data.shape)
    
    labels = append(labels, np.array([0, 1, 0, 0, 0]))
    print(labels.shape)
    
for path in fring_paths:
    wav_file, sr = librosa.load(path)
    mfcc = da.wav_to_mfcc(np.array(wav_file, dtype=float))
    
    data = append(data, mfcc)
    print(data.shape)
    
    labels = append(labels, np.array([0, 0, 1, 0, 0]))
    print(labels.shape)
    
for path in parus_paths:
    wav_file, sr = librosa.load(path)
    mfcc = da.wav_to_mfcc(np.array(wav_file, dtype=float))
    
    data = append(data, mfcc)
    print(data.shape)
    
    labels = append(labels, np.array([0, 0, 0, 1, 0]))
    print(labels.shape)
    
for path in pass_paths:
    wav_file, sr = librosa.load(path)
    mfcc = da.wav_to_mfcc(np.array(wav_file, dtype=float))
    
    data = append(data, mfcc)
    print(data.shape)
    
    labels = append(labels, np.array([0, 0, 0, 0, 1]))
    print(labels.shape)
```

```python
def append(list, element):
    return np.concatenate((list, [element]), axis=0) if list is not None else [element]
```
